### 00. Import the needed libraries and Read the dataset

### 7. Manage Duplicated Data
- Check for duplicates.
- Count the number of duplicates.
- Display the duplicated values.

```python
print("number of duplicated date:- ",file.duplicated().sum())
file[file.duplicated()]
```

```python
file.drop_duplicates(inplace=True)
file
```

### 8. Get Overall Statistics About The DataFrame

```python
file.describe().astype("int64")
```

```python
file.sum(numeric_only=True).astype("int64").to_frame(name=("Value"))
```

### 9. Display Title of The Movie Having Runtime >= 180 Minutes

```python
file[file["Runtime (Minutes)"] >= 180]
```

### 10. In Which Year There Was The Highest Average Voting?

```python
int(file["Year"].iloc[int(file["Votes"].sort_values(ascending=False).index[0])])
```

```python
int(file["Year"].iloc[file["Votes"].idxmax(skipna=True)])
```

### 11. In Which Year There Was The Highest Average Revenue?

```python
everyyear = file.groupby(by="Year")["Revenue (Millions)"].mean()

everyyear
```

```python
print(f"Highest year:- \t {everyyear.idxmax()}")
```

### 12. Find The Average Rating For Each Director

```python
file.groupby("Director")["Rating"].mean().sort_values(ascending=False)
```

### 13. Display Top 10 Lengthy Movies Title and Runtime

### 14. Display Number of Movies Per Year

### 15. Find Most Popular Movie Title (Higest Revenue)

```python
file.sort_values(by=["Revenue (Millions)"],ascending=False)
```

### 16. Display Top 10 Highest Rated Movie Titles And its Directors

### 17. Display Top 10 Highest Revenue Movie Titles

### 18.  Find Average Rating of Movies Year Wise


### 19. Count Number of Action Movies
